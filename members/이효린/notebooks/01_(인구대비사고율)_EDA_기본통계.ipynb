```python
import pandas as pd

# 파일 불러오기
df_raw = pd.read_excel("C:/Users/윤후/Downloads/2017-2021 OECD 국가 교통사고 비교.xls", header=[0, 1])
df_raw.columns = [f"{col[0]}_{col[1]}" if 'Unnamed' not in col[1] else col[0] for col in df_raw.columns]

# 1. 인구수 열 처리: 쉼표, 소수점 제거 후 숫자로 변환
population_cols = [col for col in df_raw.columns if '인구수' in col]
for col in population_cols:
    df_raw[col] = df_raw[col].astype(str).str.replace(",", "").str.replace(".", "")
    df_raw[col] = pd.to_numeric(df_raw[col], errors='coerce')

# 2. 사고, 사망 열도 숫자로 변환
accident_cols = [col for col in df_raw.columns if '사고' in col and '건' in col]
death_cols = [col for col in df_raw.columns if '사망' in col and '명' in col]
for col in accident_cols + death_cols:
    df_raw[col] = pd.to_numeric(df_raw[col], errors='coerce')

# 3. 사고율 및 사망률 계산
for year in ['2017', '2018', '2019', '2020', '2021']:
    pop_col = f'{year}_인구수(천명)'
    acc_col = f'{year}_사고[건]'
    death_col = f'{year}_사망[명]'

    if pop_col in df_raw.columns and acc_col in df_raw.columns:
        df_raw[f'{year}_사고율'] = df_raw[acc_col] / (df_raw[pop_col] / 1000)

    if pop_col in df_raw.columns and death_col in df_raw.columns:
        df_raw[f'{year}_사망률'] = df_raw[death_col] / (df_raw[pop_col] / 1000)

# 결과 확인
df_raw.head()
```
