```python
import pandas as pd
import numpy as np

# 0. Excel 파일에서 DataFrame 불러오기 (파일 경로는 사용자 경로에 맞게 수정)
df_raw = pd.read_excel('C:/Users/윤후/Downloads/2012-2016OECD 국가 교통사고 비교.xls', header=[0, 1])

# 0-1. 컬럼 이름 정리 (예: '2017', '인구수' → '2017_인구수')
df_raw.columns = [f"{col[0]}_{col[1]}" if 'Unnamed' not in col[1] else col[0] for col in df_raw.columns]

# 1. 자동차 등록대수 컬럼 추출 및 단위 변환 (천대 → 대)
car_cols = [col for col in df_raw.columns if '자동차' in col and '대수' in col]
df_raw[car_cols] = df_raw[car_cols].apply(pd.to_numeric, errors='coerce') * 1000

# 2. 인구수 컬럼 추출 및 단위 변환 (천명 → 명)
pop_cols = [col for col in df_raw.columns if '인구수' in col]
df_raw[pop_cols] = df_raw[pop_cols].apply(pd.to_numeric, errors='coerce') * 1000

# 3. 결과 확인
from IPython.display import display
display(df_raw.head())
```

```python
# 자동차 등록대수와 인구수가 모두 있는 연도별 컬럼 쌍 찾기
car_ratio_dict = {}  # 계산된 결과 저장용

for year in range(2012, 2016):
    # 컬럼명 예시: '2017_자동차등록대수(천대)', '2017_인구수(천명)'
    car_col = [col for col in car_cols if str(year) in col]
    pop_col = [col for col in pop_cols if str(year) in col]

    if car_col and pop_col:
        car_col = car_col[0]
        pop_col = pop_col[0]
        ratio_col = f"{year}_인구대비_자동차비율"
        
        df_raw[ratio_col] = df_raw[car_col] / df_raw[pop_col]
        car_ratio_dict[year] = ratio_col

# 결과 확인 (계산된 컬럼들만 보기)
df_result = df_raw[['국가_국가'] + list(car_ratio_dict.values())]
display(df_result)
```

```python
import pandas as pd

df = df_result = df_raw[['국가_국가'] + list(car_ratio_dict.values())]
display(df_result)

output_path = '인구수_자동차등록대수_분석결과.xlsx'
df.to_excel(output_path, index=False)

print(f"✅ 저장 완료: {output_path}")
```

```python
import pandas as pd
from IPython.display import display

# 자동차등록대수 : 사고건수 비율 계산
for year in range(2012, 2016):
    car_reg_col = f"{year}_자동차등록대수(천대)"
    acc_col = f"{year}_사고[건]"

    if car_reg_col in df_raw.columns and acc_col in df_raw.columns:
        # 문자열을 숫자로 변환 (에러는 NaN으로 처리)
        car_reg = pd.to_numeric(df_raw[car_reg_col], errors='coerce')
        accidents = pd.to_numeric(df_raw[acc_col], errors='coerce')

        # 계산 후 새 컬럼 생성
        df_raw[f"{year}_자동차등록대수_사고건수비"] = car_reg / accidents

# 비율 컬럼만 따로 추출해서 보기 좋게 출력
ratio_cols = [col for col in df_raw.columns if '자동차등록대수_사고건수비' in col]
df_ratio_view = df_raw[['국가_국가'] + ratio_cols]

# 표 출력
display(df_ratio_view)
```

```python
df_ratio_view.to_excel("자동차등록대수_사고건수비_분석결과.xlsx", index=False)
print("✅ 저장 완료: 자동차등록대수_사고건수비_분석결과.xlsx")
```
