```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
a1=pd.read_csv('C:/Users/이민정/OneDrive/바탕 화면/mid project/음주운전 관련 데이터/가해자_음주정도별_주야별_교통사고_2019-2023.csv')
display(a1)
```

```python
a2 = a1[['1당음주정도별(1)', '1당음주정도별(2)','주야별(1)','2019','2020','2021','2022','2023']]
display(a2)
```

```python
a2.drop(index=1)
```

```python
a2 = a1[1:].copy()

year_cols = ['2019','2020','2021','2022','2023']

a2[year_cols] = a2[year_cols].apply(pd.to_numeric, errors='coerce')

# 주야별 사고건수를 합치기 (1당음주정도별(2) 기준)
result = a2.groupby('1당음주정도별(2)')[year_cols].sum().reset_index()

# 결과 보기
print(result)
```

```python
result.to_excel('혈중알콜농도별음주사고 2019-2023.xlsx', index=False)
```
