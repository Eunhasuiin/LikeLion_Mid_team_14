```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False

from scipy import stats
```

```python
a1 = pd.read_csv('음주운전 관련 데이터/가해자_음주정도별_주야별_교통사고_2005-2018.csv')
display(a1)
```

```python
a2 = a1[['1당음주정도별(1)','1당음주정도별(2)','주야별(1)','2014','2015','2016','2017','2018']]

# 연도 열 목록
year_cols = ['2014','2015','2016','2017','2018']

# 혈줄알콜농도를 모두 합쳐준다
subset = a2.iloc[2:17].copy()

# 숫자형으로 변환
subset[year_cols] = subset[year_cols].apply(pd.to_numeric, errors='coerce')

# 주, 야 두개로 나눠 모두 합쳐주고 출력한다.
result = subset.groupby('주야별(1)')[year_cols].sum().reset_index()
display(result)
```

```python
result.to_excel('가해자_음주정도별_주야별_교통사고_2014-2018_정리.xlsx', index=False)
```
