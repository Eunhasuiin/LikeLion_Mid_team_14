```python
import pandas as pd
import numpy as np
```

```python
a1 = pd.read_excel('C:/Users/이민정/OneDrive/바탕 화면/총인구수와연간음주율병합.xlsx')
display(a1)
```

```python
for col in a1.columns:
    if col.endswith('_x'):
        base = col[:-2]
        y_col = base + '_y'
        new_col = base + '_xy'
        if y_col in a1.columns:
            a1[new_col] = np.nan
            a1.loc[:, new_col] = a1[col] * a1[y_col]

a1['19~29세_xy'] = np.nan
a1.loc[:, '19~29세_xy'] = a1['19~29세'] * a1['20~29세']

display(a1)
```

```python
a1.drop(a1.columns[1:13], axis=1, inplace=True)

a1.columns= ['연도','30~39세','40~49세','50~59세','60~69세','70세이상','19~29세']

display(a1)
```

```python
a1.to_excel('연간음주율과총인구.xlsx', index = False)
```
