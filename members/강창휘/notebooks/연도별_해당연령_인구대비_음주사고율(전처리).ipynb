```python
import pandas as pd
```

```python
age_drunk_accident = pd.read_excel('data/연령별음주사고.xlsx')
```

```python
# 구분은 모두 음주사고 -> 제거
age_drunk_accident.drop('구분', axis = 1, inplace = True)
```

```python
age_drunk_accident.head(3)
```

```python
# 데이터프레임의 구조를 변경한다.
age_drunk_accident = age_drunk_accident.melt(
    id_vars = '연도',
    value_vars = ['20세 이하', '21-30 세', '31-40 세', '41-50 세', '51-60 세', '61세 이상']
)
age_drunk_accident.columns = ['연도', '연령대', '음주사고수']
age_drunk_accident['연령대'] = age_drunk_accident['연령대'].str.replace('20세 이하', '10대')
age_drunk_accident['연령대'] = age_drunk_accident['연령대'].str.replace('21-30 세', '20대')
age_drunk_accident['연령대'] = age_drunk_accident['연령대'].str.replace('31-40 세', '30대')
age_drunk_accident['연령대'] = age_drunk_accident['연령대'].str.replace('41-50 세', '40대')
age_drunk_accident['연령대'] = age_drunk_accident['연령대'].str.replace('51-60 세', '50대')
age_drunk_accident['연령대'] = age_drunk_accident['연령대'].str.replace('61세 이상', '60대이상')
```

```python
# 정렬 기준을 연도와 연령대에 맞게 변경하고 인구 데이터의 연도에 맞게 수정한다.
age_drunk_accident = age_drunk_accident.sort_values(['연도', '연령대']).reset_index(drop = True)
age_drunk_accident['연령대'] = age_drunk_accident['연령대'].str.replace('10대', '20대이하')
age_drunk_accident = age_drunk_accident.loc[age_drunk_accident['연도'] >= 2015].reset_index(drop = True)
```

```python
age_population = pd.read_excel('data/총인구수2015~2021.xlsx')
```

```python
#행정구역, 항목은 칼럼 내 모든 데이터가 같음 -> 제거
age_population.drop(['행정구역별(읍면동)', '항목'], axis = 1, inplace = True)
```

```python
age_population = age_population.loc[age_population['연령별'].isin(['20대', '30대', '40대', '50대', '60대', '70대이상'])].reset_index(drop = True)
```

```python
# 연령별 칼럼이 60대, 70대이상인 행을 모두 더해서 새로운 칼럼을 만든다. 
age_population.loc[6, :] = age_population.loc[4:].sum()
```

```python
# 60대, 70대 이상이 있는 인덱스를 삭제한다.
age_population = age_population.drop([4, 5]).reset_index(drop = True)
```

```python
age_population.loc[4, '연령별'] = '60대이상'
```

```python
# 데이터프레임 구조 변경
age_population = age_population.melt(
    id_vars = '연령별',
    value_vars = ['2015 년', '2016 년', '2017 년', '2018 년', '2019 년', '2020 년', '2021 년']
)
# 칼럼 변경
age_population.columns = ['연령대', '연도', '인구수']
# 순서 변경
age_population = age_population[['연도', '연령대', '인구수']]
```

```python
# 인구수를 정수 타입으로 변경
age_population['인구수'] = age_population['인구수'].astype('int')
```

```python
# " 년" 문자열 제거
age_population['연도'] = age_population['연도'].str.replace('2015 년', '2015')
age_population['연도'] = age_population['연도'].str.replace('2016 년', '2016')
age_population['연도'] = age_population['연도'].str.replace('2017 년', '2017')
age_population['연도'] = age_population['연도'].str.replace('2018 년', '2018')
age_population['연도'] = age_population['연도'].str.replace('2019 년', '2019')
age_population['연도'] = age_population['연도'].str.replace('2020 년', '2020')
age_population['연도'] = age_population['연도'].str.replace('2021 년', '2021')
```

```python
age_population2 = pd.read_excel('data/총인구수(추가).xlsx')
```

```python
# 필요없는 칼럼 및 행 삭제
age_population2 = age_population2.drop([0, 1]).reset_index(drop = True)
age_population2.drop('행정구역별(읍면동)', axis = 1, inplace = True)
```

```python
# 본격적인 전처리 시작
age_population2 = age_population2.iloc[:4, :]
# 모든 행을 모두 더하고 20대이하 행을 만들어 준다.
age_population2.loc[4, :] = age_population2.sum()
age_population2 = age_population2.drop([0, 1, 2, 3]).reset_index(drop = True)
age_population2.loc[0,'연령별'] = '10대'
```

```python
# 데이터프레임 구조 변경
age_population2 = age_population2.melt(id_vars = '연령별',
                                       value_vars = ['2015', '2016', '2017', '2018', '2019', '2020', '2021']
                                      )
# 칼럼 변경
age_population2.columns = ['연령대', '연도', '인구수']
# 순서 변경
age_population2 = age_population2[['연도', '연령대', '인구수']]
```

```python
age_population_c = pd.concat([age_population, age_population2])
```

```python
age_population_c = age_population_c.sort_values(['연도', '연령대']).reset_index(drop = True)
age_population_c['연령대'] = age_population_c['연령대'].str.replace('10대', '20대이하')
age_population_c['연도'] = age_population_c['연도'].astype('int')
age_population_c['인구수'] = age_population_c['인구수'].astype('int')
```

```python
drunk_accident_population = age_drunk_accident.copy()
drunk_accident_population['인구수'] = age_population_c['인구수']
```

```python
drunk_accident_population.to_csv('연도별_연령별_인구대비_음주사고수(전처리).csv', index = False)
```
