```python
import pandas as pd
```

```python
d_rate = pd.read_excel('data/연간음주율.xlsx')
```

```python
# 필요없는 칼럼 제거 (index, 구분, 전체)
d_rate = d_rate.drop(['Unnamed: 0', '구분', '전체'], axis = 1)
d_rate.columns = ['연도', '20대', '30대', '40대', '50대', '60대', '70대이상']
```

$$
\text{60대 이상 비율} = \frac{(\text{60대 비율} \times \text{60대 인구 수}) + (\text{70대 이상 비율} \times \text{70대 이상 인구 수})}{\text{60대 인구 수} + \text{70대 이상 인구 수}}
$$

```python
# 이전 데이터와 호환 시키기 위해서는 60대이상의 음주율을 알아야한다. 
d_rate
```

```python
# 60대, 70대 이상 인구 데이터를 구해온다.
age_drunk_60_70 = pd.read_csv('data(전처리)/60_70_population.csv')
```

```python
# 60대, 70대 이상 인구 데이터를 나눈다.
age_drunk_60 = age_drunk_60_70.loc[age_drunk_60_70['연령대'] == '60대'].reset_index(drop = True)
age_drunk_70 = age_drunk_60_70.loc[age_drunk_60_70['연령대'] == '70대이상'].reset_index(drop = True)
```

```python
# 60대이상 음주율 계산
d_rate_60up = ((d_rate['60대'] * age_drunk_60['인구수']) + (d_rate['70대이상'] * age_drunk_70['인구수'])) / (age_drunk_60['인구수'] + age_drunk_70['인구수'])
d_rate_60up = round(d_rate_60up, ndigits = 1)
d_rate_60up
```

```python
# 60대 이상 음주율을 칼럼에 추가
d_rate['60대이상'] = d_rate_60up
# 60대, 70대 이상 음주율을 d_rate에서 제거한다.
d_rate = d_rate.drop(['60대', '70대이상'], axis = 1)
```

```python
d_rate
```

```python
# 20대 이하의 음주율 가져오기
minor_d_rate = pd.read_csv('data/연도별_청소년음주율.csv')
```

```python
# 20대 이하 음주율 계산
minor_d_rate = minor_d_rate.loc[10:16, :'전체'].reset_index(drop = True)
```

```python
minor_d_rate['전체']
```

```python
# d_rate에 20대 이하 음주율 추가
d_rate['10대'] = minor_d_rate['전체']
d_rate = d_rate[['연도', '10대', '20대', '30대', '40대', '50대', '60대이상']]
```

```python
d_rate
```

```python
d_rate = d_rate.melt(id_vars = '연도', value_vars = ['10대', '20대', '30대', '40대', '50대', '60대이상'])
d_rate.columns = ['연도', '연령대', '음주율']
```

```python
d_rate = d_rate.sort_values(['연도', '연령대']).reset_index(drop = True)
```

```python
d_rate
```
